```python
import pandas
import numpy
from datetime import date
```

```python
station_data = pandas.read_csv('./cycle-dataset/station.csv')
trip_data = pandas.read_csv('./cycle-dataset/trip.csv', skiprows=[50793])

weather_data = pandas.read_csv('./cycle-dataset/weather.csv')
# station_data.head()
# weather_data.head()

```

```python
# trip_data.head()
```

### i) What is the average trip duration for a borrowed bicycle?

```python
duration_avg = round(trip_data['tripduration'].mean(), 2)
print('average trip duration: {} seconds'.format(duration_avg))
```

### ii) What's the most common age of a bicycle-sharer?

```python
trip_data['starttime'] = pandas.to_datetime(trip_data['starttime'], errors='coerce')
trip_data['age'] = trip_data['starttime'].dt.year - trip_data['birthyear']
common_age = trip_data['age'].mode()
print('most common rider age: {} years old.'.format(common_age))
# trip_data.head()
```

### iii) Given all the weather data here, find the average precipitation per month, and the median precipitation.


```python
# weather_data.head()
```

```python
weather_data['parsedate'] = pandas.to_datetime(weather_data['Date'], errors='coerce')
weather_data['month'] = weather_data['parsedate'].dt.month
rain_by_month = weather_data[['month', 'Precipitation_In']].groupby(weather_data.month)
sum_prcp = pandas.DataFrame(rain_by_month['Precipitation_In'].sum())
sum_prcp
```

### iv) Whatâ€™s the average number of bikes at a given bike station?


```python
# station_data.head()
```

```python
avg_bikes = station_data[['name', 'current_dockcount']].groupby(station_data.name)
avg_bikes.head()
```

### v) When a bike station is modified, is it more likely that itâ€™ll lose bikes or gain bikes? How do you know?


```python
station_data['after_mod_count'] = station_data['install_dockcount'] - station_data['current_dockcount']
print('average increase after mod: {} bikes'.format(round(station_data['after_mod_count'].mean(), 2)))
```
